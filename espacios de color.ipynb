```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('imagenes/car.jpg')
cv.imshow('Auto', img)

#BGR a RGB en matplotlib
# plt.imshow(img)
# plt.show()

# #convirtiendo una imagen a escala de grises
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('Gray', gray)

#convirtiendo una imagen a HSV
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow('HSV', hsv)

#convirtiendo una imagen a L*a*b
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)
cv.imshow('LAB', lab)

#BGR a RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
cv.imshow('RGB', rgb)

plt.imshow(rgb)
plt.show()

cv.waitKey(0)
```

No se puede convertir de hsv directamente a grayscale
